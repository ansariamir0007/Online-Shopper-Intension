```python
sales=shop[shop['Revenue']==True]
sales=pd.get_dummies(sales,columns=['Month','VisitorType']).reset_index().drop('index',axis=1)
```

```python
sales=sales.drop('Revenue',axis=1)
```

```python
sales.info()
```

```python
from sklearn.cluster import KMeans
ssd=[]
for num_cluster in list(range(1,21)):
    model_clus=KMeans(n_clusters=num_cluster,max_iter=50)
    model_clus.fit(sales)
    ssd.append(model_clus.inertia_)
plt.plot(ssd)
```

```python
model_clus=KMeans(n_clusters=5)
model_clus.fit(sales)
```

```python
model_clus.labels_
```

```python
sales['clust_id']=model_clus.labels_
```

```python
plt.subplots(figsize=(15,5))
sns.scatterplot('PageValues','ExitRates',hue='clust_id',data=sales,palette='summer_r')
```

```python
plt.subplots(figsize=(15,5))
sns.scatterplot('ProductRelated','ProductRelated_Duration',hue='clust_id',data=sales,palette='summer_r')
```
